```python
import pandas as pd
import numpy as np
```

```python
file_path_games = '/Users/bobbyzhu/Desktop/academic/DSC/dsc106/final_project/data/games.csv'
file_path_ranking = '/Users/bobbyzhu/Desktop/academic/DSC/dsc106/final_project/data/ranking.csv'
file_path_teams = '/Users/bobbyzhu/Desktop/academic/DSC/dsc106/final_project/data/teams.csv'

games_df = pd.read_csv(file_path_games)
teams_df = pd.read_csv(file_path_teams)
games_df
```

```python

nba_teams_nicknames_states = {
    "Hawks": "Georgia",
    "Celtics": "Massachusetts",
    "Nets": "New York",
    "Hornets": "North Carolina",
    "Bulls": "Illinois",
    "Cavaliers": "Ohio",
    "Mavericks": "Texas",
    "Nuggets": "Colorado",
    "Pistons": "Michigan",
    "Warriors": "California",
    "Rockets": "Texas",
    "Pacers": "Indiana",
    "Clippers": "California",
    "Lakers": "California",
    "Grizzlies": "Tennessee",
    "Heat": "Florida",
    "Bucks": "Wisconsin",
    "Timberwolves": "Minnesota",
    "Pelicans": "Louisiana",
    "Knicks": "New York",
    "Thunder": "Oklahoma",
    "Magic": "Florida",
    "76ers": "Pennsylvania",
    "Suns": "Arizona",
    "Trail Blazers": "Oregon",
    "Kings": "California",
    "Spurs": "Texas",
    "Raptors": "Ontario",  # This is a province, not a state.
    "Jazz": "Utah",
    "Wizards": "District of Columbia"  
}


games_df = games_df[['GAME_DATE_EST','GAME_ID','HOME_TEAM_ID','VISITOR_TEAM_ID','SEASON','HOME_TEAM_WINS']]
teams_df = teams_df[['TEAM_ID','CITY','NICKNAME']]
games_df = games_df.merge(teams_df,left_on='HOME_TEAM_ID',right_on='TEAM_ID').rename(columns={'CITY': 'HOME_CITY'}).rename(columns={'NICKNAME': 'HOME_NICKNAME'}).drop(columns=['TEAM_ID'])
games_df = games_df.merge(teams_df,left_on='VISITOR_TEAM_ID',right_on='TEAM_ID').rename(columns={'CITY': 'VISITOR_CITY'}).rename(columns={'NICKNAME': 'VISITOR_NICKNAME'}).drop(columns=['TEAM_ID'])
games_df = games_df.sort_values(by='GAME_DATE_EST',ascending=False).reset_index().drop(columns=['index'])
games_df['HOME_STATE'] = games_df['HOME_NICKNAME'].apply(lambda x: nba_teams_nicknames_states[x])
games_df
```

```python
games_df.to_csv('plot_1_data.csv')
```
